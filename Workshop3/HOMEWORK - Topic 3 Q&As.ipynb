```python
# Answer 1:
def calculate_cylinder_volume(radius, height):
    volume = 3.14 * radius**2 * height
    return volume

# Example usage:
cylinder_volume = calculate_cylinder_volume(2, 5)
print("Cylinder Volume:", cylinder_volume)
```

```python
# Topic: Create a Simple Class

# Question 5: Create a class named 'Book' with attributes 'title' and 'author' initialized in the __init__ method.
# Your code here
```

```python
# Answer 5:
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

# Example usage:
my_book = Book("The Great Gatsby", "F. Scott Fitzgerald")
print("Book Title:", my_book.title)
print("Book Author:", my_book.author)
```

```python
# Question 6: Create a class named 'Dog' with methods 'bark' and 'fetch'.
# Your code here
```

```python
# Answer 6:
class Dog:
    def bark(self):
        print("Woof! Woof!")

    def fetch(self):
        print("Fetching the ball!")

# Example usage:
my_dog = Dog()
my_dog.bark()
my_dog.fetch()
```
