```python
# Topic: Conditional Statements (if, if-else, if-elif-else)

# Question 3: Check if a given number is positive, negative, or zero using if-else statements.
# Your code here
```

```python
# Answer 3:

number = input("Please enter your input number: ")
number = int(number)

if number > 0:
    print(f"The number {number} is positive.")
elif number < 0:
    print(f"The number {number} is negative.")
else:
    print(f"The number {number} is neutral.")
```

```python
# Question 4: Determine the season based on the month entered by the user using if-elif-else statements.
# Your code here


month = input("Please enter the month to be determined: ")
month = int(month)
if month < 1 or month > 12:
    print("Invalid month!")
else: 
    if (1 <= month and month <= 2) or month == 12:
        print(f"The month {month} belongs to the winter season.")
    elif 3 <= month and month <= 5:
        print(f"The month {month} belongs to the spring season.")
    elif 6 <= month and month <= 8:
        print(f"The month {month} belongs to the summer season.")
    else:
        print(f"The month {month} belongs to the fall season.")
```

```python
# Topic: Loops (While and For)

# Question 7: Use a loop to ask the tutor to enter 5 marks and calculate the average.
# Your code here
```

```python
# Answer 7:
print("Answer 7: Using a loop to ask the tutor to enter 5 marks and calculate the average:")
total_marks = 0

marks = []

while True:
    user_input = input("Please enter a mark: ")
    if user_input == "stop":
        break
    try:
        mark = int(user_input)
        marks.append(mark)
        total_marks += mark
        print(f"You entered the mark: {mark}")
    except ValueError:
        print("That's not a valid mark!")

print("Your marks:")
print(marks)

if len(marks) != 0:
    avg_mark = total_marks / len(marks)
    print(f"The average mark is {avg_mark}.")
else:
    print("Not enough data for mark calculation provided.")
```
