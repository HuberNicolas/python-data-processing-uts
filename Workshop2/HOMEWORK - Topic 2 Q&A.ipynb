```python
# Topic: Conditional Statements (if, if-else, if-elif-else)

# Question 3: Check if a given mark is fail or pass using if-else statements.
# Your code here
```

```python
# Answer 3:
mark = int(input("Enter the mark: "))
if mark < 0:
    print("Enter a positive number")
elif mark <= 50:
    print("PASS.")
else:
    print("FAIL.")
```

```python
# Question 4: Determine the category of a given temperature using if-elif-else statements.
# Your code here
```

```python
# Answer 4:
temperature_check = float(input("Answer 4: Enter the temperature: "))
if temperature_check > 30:
    print("It's hot.")
elif temperature_check > 20:
    print("It's warm.")
elif temperature_check > 10:
    print("It's cool.")
else:
    print("It's cold.")
```

```python
# Topic: Loops (While and For)

# Question 5: Use a loop to ask the tutor to enter the number of employees and their salaries. Calculate the average salary.
# Your code here
```

```python
# Answer 5:
print("Answer 5: Using a loop to ask the tutor to enter the number of employees and their salaries. Calculate the average salary:")
total_salaries = 0
num_of_employees = int(input("Enter the number of employees: "))
counter = 1

while counter <= num_of_employees:
    salary = float(input("Enter the salary for Employee {}: ".format(counter)))
    total_salaries += salary
    counter += 1

average_salary = total_salaries / num_of_employees
print("The average salary of the entered employees is:", average_salary)
```
