# Scenario: Analyzing Telecom Churn

Question: After reading the Orange Telecom's Churn Dataset, which consists of cleaned customer activity data, let's delve into the statistics to tell a comprehensive story about customer churn.

**What is the overall churn count in the dataset?**

**What percentage of customers have an international plan or a voice mail plan?**

```python
# Display counts of Churn values (True and False)
#code here
df['churn'].value_counts()
```

```python
#What percentage of customers have an international plan or a voice mail plan?
#code here
# ???
international_plan_percentage = df['international plan'].value_counts()
print(international_plan_percentage)
voice_mail_plan_percentage = df['voice mail plan'].value_counts()
print(voice_mail_plan_percentage)
```

Customer Service Interaction:

How many customer service calls do customers typically make?

Is there a correlation between the number of customer service calls and churn?


```python
# Analyze customer service calls and its relation to churn
#code here
# Analyze customer service calls and its relation to churn
average_service_calls = df['customer service calls'].mean()
print(average_service_calls)
```

```python
#Is there a correlation between the number of customer service calls and churn?
#code here

service_calls_churn_correlation = df['customer service calls'].corr(df['churn'])
service_calls_churn_correlation
```

Usage Patterns:

What are the average total day, evening, and night charges for customers?

Is there a noticeable difference in usage patterns between customers who churn and those who don't?

```python
# What are the average total day, evening, and night charges for customers?
average_day_charge = df['total day charge'].mean()
average_eve_charge = df['total eve charge'].mean()
average_night_charge = df['total night charge'].mean()

#Is there a noticeable difference in usage patterns between customers who churn and those who don't?
churn_usage_stats = df.groupby('churn')[['total day minutes', 'total eve minutes', 'total night minutes']].mean()
churn_usage_stats
```

International Usage:

What is the average number of international minutes for customers who churn compared to those who don't?

Is having an international plan associated with a lower churn rate?

```python
#What is the average number of international minutes for customers who churn compared to those who don't?
average_intl_minutes_churn = df.groupby('churn')['total intl minutes'].mean()
print(average_intl_minutes_churn)

#Is having an international plan associated with a lower churn rate?
intl_plan_churn_rate = df.groupby('international plan')['churn'].value_counts()
print(intl_plan_churn_rate)
```
