## Import Matplotlib and Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Question 1:
# Create a bar plot for total rental bikes based on workingday
plt.figure(figsize=(10, 6))
sns.barplot(x='workingday', y='total rental bikes', data=df, ci=None, palette='viridis')
plt.title('Total Rental Bikes on Working Days vs Non-Working Days')
plt.xlabel('Working Day (1: Yes, 0: No)')
plt.ylabel('Total Rental Bikes')
plt.show()
```

```python
#Question 2:
 # Create a bar plot for total rental bikes based on weekday
plt.figure(figsize=(10, 6))
sns.barplot(x='weekday', y='total rental bikes', data=df, ci=None, palette='viridis')
plt.title('Total Rental Bikes on weekday')
plt.xlabel('Week days')
plt.ylabel('Total Rental Bikes')
plt.show()
```

```python
#Question 3:
# Select numerical attributes for pairplot
numerical_attributes = df[['temp normalized', 'original hum', 'windspeed', 'casual users', 'registered users', 'total rental bikes']]

# Create a pairplot
sns.pairplot(numerical_attributes)
plt.suptitle("Pairplot of Numerical Attributes", y=1.02)
plt.show()
```

```python
#Question 4:
# Select numerical attributes for the heatmap
numerical_attributes = df[['temp normalized', 'original hum', 'windspeed', 'casual users', 'registered users', 'total rental bikes']]

# Calculate the correlation matrix
correlation_matrix = numerical_attributes.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap of Numerical Attributes")
plt.show()
```

```python
#Question 5:Create pie chart weathersit
# Count the occurrences of each weather situation
weather_counts = df['weathersit'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title("Distribution of Weather Situations")
plt.show()
```
