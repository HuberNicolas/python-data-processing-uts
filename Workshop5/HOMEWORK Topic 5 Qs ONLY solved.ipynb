```python
#Question 12: Prompt the user to enter details for multiple products, including product name, price, and quantity.
#Store this information in a list of dictionaries representing each product. Finally, display the list of products.
```

```python
products_list = []
num_products = int(input("Enter the number of products: "))


for i in range(num_products):
    
    prod_dict = {"name": "", "price": 0.0, "quantity": 0}
    product_name = str(input(f"For product {i}, enter product name: "))
    product_price = float(input(f"For product {i}, enter product price: "))
    product_quantity = int(input(f"For product {i}, enter product quantity: "))
    
    prod_dict["name"] = product_name
    prod_dict["price"] = product_price
    prod_dict["quantity"] = product_quantity
    
    products_list.append(prod_dict)
    
    
for product in products_list:
    print(product)
```

```python
# Question 13 Collect exam scores for a group of students using the input function. Add the scores to a dictionary,
#and then categorize each student based on their scores into predefined grade categories (A, B, C, D).
```

```python
exam_scores = {}
num_students = int(input("Enter the number of students: "))



for i in range(num_students):
    grade = str(input(f"Enter grade {i}: "))
    
    if grade in exam_scores.keys():
        exam_scores[grade] += 1
    
    else:
        exam_scores[grade] = 1


a_count = exam_scores["A"]
b_count = exam_scores["B"]
c_count = exam_scores["C"]
d_count = exam_scores["D"]

print("Grade Categories:")
print("A:", a_count)
print("B:", b_count)
print("C:", c_count)
print("D:", d_count)
```
