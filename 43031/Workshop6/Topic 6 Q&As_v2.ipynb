Question 3:
Access elements of a 2D array in Python. Retrieve and print the first and last element.
Retrieve and print the first and last column

```python
# Find the temperature of SYD
first_element = my_2d_array[0][1]
print("SYD temp:", first_element)

# Retrieve and print the last element
last_element = my_2d_array[-1][-1]
print("Last element:", last_element)

# Retrieve and print the first column
first_column=[]
for row in my_2d_array:
  first_column.append(row[0])
print("First column:", first_column)

# Retrieve and print the last column
last_column=[]
for row in my_2d_array:
  last_column.append(row[-1])
print("Last column:", last_column)
```

Question 4:
Use the open function and csv library to read all the content of a csv file (iris.csv) in Python. Print the content of the file.

```python
import csv
csv_file_path = '/content/drive/MyDrive/430031 notebooks/Week 6/iris.csv'
# Open the CSV file for reading
with open(csv_file_path, 'r') as csvfile:
  csv_reader = csv.reader(csvfile)
  data = list(csv_reader)
print(data)
```

Question 7:  Calculates the average of columns "sepal.length","sepal.width","petal.length","petal.width"

```python
total_columns=[0,0,0,0]
data_wout_header=data[1:]

for line in data_wout_header:
  for i in range(4):
    total_columns[i]=total_columns[i] + float(line[i])
print("Total",total_columns)

for val in total_columns:
  print(val/len(data_wout_header))
```
