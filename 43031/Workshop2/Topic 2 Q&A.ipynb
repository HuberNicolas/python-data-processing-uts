```python
# Topic: Conditional Statements (if, if-else, if-elif-else)

# Question 3: Check if a given number is positive, negative, or zero using if-else statements.
# Your code here
```

```python
# Answer 3:
number = int(input("Answer 3: Enter a number: "))
if number > 0:
    print("The number is positive.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
# Question 4: Determine the season based on the month entered by the user using if-elif-else statements.
# Your code here
```

```python
# Answer 4:
month = input("Answer 4: Enter a month: ").lower()

if month == "december" or month == "january" or month == "february":
    print("It's winter.")
elif month == "march" or month == "april" or month == "may":
    print("It's spring.")
elif month == "june" or month == "july" or month == "august":
    print("It's summer.")
else:
    print("It's fall.")
```

```python
# Topic: Loops (While and For)

# Question 5: Use a while loop to print numbers from 1 to 5.
# Your code here
```

```python
# Answer 5:
print("Answer 5: Using a while loop to print numbers from 1 to 5:")
counter = 1
while counter <= 5:
    print(counter, end=" ")
    counter += 1
print()
```

```python
# Question 6: Use a for loop to print numbers from 1 to 5.
# Your code here, search about range function
```

```python
# Answer 6:
print("Answer 6: Using a for loop to iterate through a range:")
for num in range(1, 6):
    print(num, end=" ")
```

```python
# Topic: Loops (While and For)

# Question 7: Use a loop to ask the tutor to enter 5 marks and calculate the average.
# Your code here
```

```python
# Answer 7:
print("Answer 7: Using a loop to ask the tutor to enter 5 marks and calculate the average:")
total_marks = 0

for i in range(1, 6):
    mark = float(input("Enter mark {}: ".format(i)))
    total_marks += mark

average = total_marks / 5
print("The average of the entered marks is:", average)
```
