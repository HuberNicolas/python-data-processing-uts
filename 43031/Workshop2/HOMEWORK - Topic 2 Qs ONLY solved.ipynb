```python
# Question 4: Determine the category of a given temperature using if-elif-else statements.
# Your code here
```

```python
# Answer 4:

temp = input("Please enter the current temperature: ")
temp = int(temp)

if temp < 10:
    print("Its a bit chilly today.")
elif temp >= 10 and temp < 20:
    print("Its not warm today")
else:
    print("Today its quite warm.")
```

```python
# Topic: Loops (While and For)

# Question 5: Use a loop to ask the tutor to enter the number of employees and their salaries. Calculate the average salary.
# Your code here

```

```python
# Answer 5:
print("Answer 5: Using a loop to ask the tutor to enter the number of employees and their salaries. Calculate the average salary:")



salaries = []

while True:
    user_input = input("Please enter a salary: ")
    if user_input == "stop":
        break
    try:
        salary = int(user_input)
        salaries.append(salary)
        print(f"You entered the salary: {salary}")
    except ValueError:
        print("That's not a valid salary!")

print("Your salaries:")
print(salaries)

if len(salaries) != 0:
    avg_salary = sum(salaries) / len(salaries)
    print(f"The average salary is {avg_salary}.")
else:
    print("Not enough data for salary calculation provided.")
```
