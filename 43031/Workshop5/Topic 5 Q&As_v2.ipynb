```python
#Question 11: Use the input function to get salaries from users, add them to a list, and then calculate and print the average salary.
```

```python
# Answer 11
salary_list = []
num_salaries = int(input("Enter the number of salaries: "))

for i in range(num_salaries):
    salary = float(input("Enter salary "))
    salary_list.append(salary)

average_salary = sum(salary_list) / len(salary_list)
print("Average Salary:", average_salary)
```

```python
#to stop the code if the user enters non-numeric value, use function isinstance(x, float)
salary_list = []
num_salaries = int(input("Enter the number of salaries: "))

for i in range(num_salaries):
    while True:
        salary_input = input("Enter salary: ")
        if salary_input.replace('.', '', 1).isdigit():  # Remove "." and check if all are numbers Check if input is numeric
            salary = float(salary_input)
            break  # Exit the loop if input is numeric
        else:
            print("Error: Please enter a valid numeric value.")

    salary_list.append(salary)

average_salary = sum(salary_list) / len(salary_list)
print("Average Salary:", average_salary)
```

```python
# Question 12 Use input function to add marks to dictionary then count of the number of students who pass and fail
```

```python
# Question 12
marks_dict = {}
num_students = int(input("Enter the number of students: "))

for i in range(num_students):
    student_name = input("Enter the student name ")
    mark = int(input("Enter the mark "))
    marks_dict[student_name] = mark

# Counting grade categories
# Initialize counters for each category
P_count = 0
F_count = 0

# Assuming marks_dict contains the student names as keys and their marks as values
for mark in marks_dict.values():
    if mark >= 50:
        P_count += 1
    else:
        F_count += 1

# Print the counts for each category
print("Pass:", P_count)
print("Fail:", F_count)
```
