```python
#Question 12: Prompt the user to enter details for multiple products, including product name, price, and quantity.
#Store this information in a list of dictionaries representing each product. Finally, display the list of products.
```

```python
products_list = []
num_products = int(input("Enter the number of products: "))

for i in range(num_products):
    product_info = {}
    product_info["name"] = input(f"Enter product name {i+1}: ")
    product_info["price"] = float(input(f"Enter price for {product_info['name']}: "))
    product_info["quantity"] = int(input(f"Enter quantity for {product_info['name']}: "))
    products_list.append(product_info)

# Display the list of products
print("List of Products:")
for product in products_list:
    print(product)
```

```python
# Question 13 Collect exam scores for a group of students using the input function. Add the scores to a dictionary,
#and then categorize each student based on their scores into predefined grade categories (A, B, C, D).
```

```python
exam_scores = {}
num_students = int(input("Enter the number of students: "))

for i in range(num_students):
    student_name = input(f"Enter student name {i+1}: ")
    exam_scores[student_name] = int(input(f"Enter exam score for {student_name}: "))

# Categorizing students
grade_categories = {"A": 90, "B": 80, "C": 70, "D": 60}
student_categories = {}

for student, score in exam_scores.items():
    for grade, threshold in grade_categories.items():
        if score >= threshold:
            student_categories[student] = grade
            break

print("Students Categorized by Grades:")
for student, category in student_categories.items():
    print(f"{student}: {category}")
```
