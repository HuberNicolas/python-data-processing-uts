```python
# Answer 1:

import math

def calculate_cylinder_volume(radius, height):

    volume = math.pi * (radius**2) * height
    return volume
    

print(calculate_cylinder_volume(10, 10))
```

```python
# Question 6: Create a class named 'Dog' with methods 'bark' and 'fetch'.
# Your code here
```

```python
# Answer 6:
class Dog:
    def __init__(self):
        pass
    
    def bark(self):
        print("Wau!")
    
    def fetch(self):
        print("The the dog will fetch something.")
        
bello = Dog()
bello.bark()
bello.fetch()
```
