```python
# Answer 5:

class Car():
    def __init__(self, make, model):
        self.make = make
        self.model = model
        
    

# Example usage:
my_car = Car("Toyota", "Camry")
print("My car: {} {}".format(my_car.make, my_car.model))
```

```python
# Question 6: Create a class named 'Rectangle' with methods 'calculate_area' and 'calculate_perimeter'.
# Your code here
```

```python
# Answer 6:
class Rectangle:
    def __init__(self, width, length):
        self.width = width
        self.length = length
    
    def calculate_area(self):
        return self.width * self.length

    def calculate_perimeter(self):
        return 2*(self.width + self.length)
```

```python
rec = Rectangle(4,5)
print(f"Rectangle rec with width {rec.width} and length {rec.length} has area {rec.calculate_area()} and perimeter {rec.calculate_perimeter()}")
```
