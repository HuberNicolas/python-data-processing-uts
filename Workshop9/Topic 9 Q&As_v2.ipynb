## Missing values

```python
# Question 1: Use isnull() or info() function in Pandas to identify the missing values?
print(data.shape)
missing_values_per_column = data.isnull().sum()
print(missing_values_per_column)
```

# Data cleaning to solve data quality issues

## Handling missing values by droping rows

```python
# Question 2: Handle missing values by removing rows with missing values in a Pandas DataFrame using dropna() method?

#assign data to another variable called data_no_missing_values to avoid changes on the same dataset
data_no_missing_values = data.dropna()
data_no_missing_values.head()

#Show the stats of null values
print(data_no_missing_values.isnull().sum())
```

```python
#How many rows have been removed?
rem_rows=len(data) - len(data_no_missing_values)
print("Number of rows removed:", rem_rows)
```

## Handling missing values by data imputation


*   if the variable is numerical, replace the missing values by mean or median
*   if the variable is non-numerical (categorical), replace the missing values by most frequent value



```python
# Question 3: Handle missing values by replacing numerical missing values with the mean in a Pandas DataFrame using fillna method?

#identify the numerical variables first
numerical_columns = data.select_dtypes(include=['int', 'float']).columns

#iterate over each column and relpalce missing by mean
for column in numerical_columns:
  mean_value = data[column].mean()
  data[column].fillna(mean_value, inplace=True)

print(data.head())

#Any issues you notice after the showing the results?property_id?
```

```python
numerical_columns
```

```python
# Question 4: How do you handle missing values by replacing categorical missing values with the most frequent value in a Pandas DataFrame?

#identify the non-numeric variables first
categorical_columns = data.select_dtypes(include='object').columns

#iterate over each column and relpalce missing by mode
for column in categorical_columns:
  mode_value = data[column].mode()[0]
  data[column].fillna(mode_value, inplace=True)
print(data.head())

```

```python
#Check again the previous code to see if any missing value exist
missing_values_per_column = data.isnull().sum()

# Display the number of missing values per column
missing_values_per_column
```

## Duplications

```python
# Question 5: Find the duplicate rows using duplicated() method?

# Check for duplicate rows
duplicate_rows = data[data.duplicated()]
print(duplicate_rows.shape)
print(duplicate_rows)
```

```python
# Question 6: Find the duplicate columns using duplicated() method?

#Transpose the dataset first then call .duplicated()
duplicate_columns = data.transpose().duplicated()
print(duplicate_columns)
```

## Handling duplicates


*   Remove duplicate rows
*   Remove duplicate columns

```python
# Question 7: How to Remove duplicate rows drop_duplicates()?

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Check duplicates row
print(data[data.duplicated()])
```

```python
# Question 8: Remove duplicate columns?

#Transpose the data
df_transposed=data.transpose()
df_transposed.drop_duplicates(inplace=True)
data=df_transposed.transpose()

print(data.head())

```

**Data cleaning has been completed, and there are no more missing values or duplicates in the dataset. It is now ready for data analysis.**
